CLEANING THE TITANIC DATASET

```python
Median =Titanic['Age'].median()
Median
```

```python
# Age column has 177 empty cells. To fill the empty cells with median which is the most appropriate
Titanic['Age'].fillna(Titanic['Age'].median(), inplace=True)
```

```python
##To confirm if there are still empty cells in the age column
Titanic.isnull().sum()
```

```python
##Dropping of coulumns that we will not be using
Titanic.drop(['PassengerId', 'Name', 'Ticket', 'Cabin', 'Embarked'], axis=1, inplace=True)
Titanic
```

```python
Titanic.info()
```

Visualization to compare survival rates across independent variables

```python
# Bar plot of survival rate by Pclass
sns.barplot(x="Pclass", y="Survived", data=Titanic)
plt.title("Survival Rate by Passenger Class")
plt.show()
```

This count plot above suggests that females had a higher survival rate than males

```python
# Bar plot of survival rate by Sex
sns.barplot(x="Sex", y="Survived", data=Titanic)
plt.title("Survival Rate by Sex")
plt.show()
```

```python
# Box plot of survival rate by Age
sns.boxplot(x="Age", y="Survived", data=Titanic)
plt.title("Age Distribution by Survival")
plt.show()
```

```python
# Histogram plot of survival rate by Fare
sns.histplot(Titanic[Titanic["Survived"] == 1]["Fare"], bins=30, color="green", kde=True, label="Survived")
sns.histplot(Titanic[Titanic["Survived"] == 0]["Fare"], bins=30, color="red", kde=True, label="Did Not Survive")
plt.legend()
plt.title("Fare Distribution by Survival")
plt.show()
```

```python
#Label encode SEX column to numerial variables
from sklearn.preprocessing import LabelEncoder
#initialize LabelEncoder
le = LabelEncoder()
#Fit and transform the 'Sex' column
Titanic['Sex'] = le.fit_transform(Titanic['Sex'])
```

```python
#To confirm the data is loaded correctly
Titanic.head()
```

```python
#Converting the column 'Age' from float to integer
Titanic['Age']=Titanic['Age'].astype(int)
```

```python
Titanic.head()
```

DIVIDING TITANIC DATASET INTO INDEPENDENT AND DEPENDENT VARIBLES

```python
from sklearn.model_selection import train_test_split
X = Titanic.drop('Survived', axis=1)    
y = Titanic['Survived']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
X_test
```

Model Implementation and Evaluation
I experimented with both Decision Tree and Random Forest for predicting the survival of passengers on the Titanic to see which one generalizes better on the dataset. Below are the codes and output to compare results:

```python
from sklearn.tree import DecisionTreeClassifier

# Train Decision Tree Model
model= DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
Predictions = model.predict(X_test)
Predictions
```

```python
#Evaluation Metrics
from sklearn.metrics import accuracy_score
score= accuracy_score(y_test, Predictions)
score
```

```python
#Using another machine learning model to see if it is better than the previous model
from sklearn.ensemble import RandomForestClassifier

# Train Random Forest Model
model_1 = RandomForestClassifier()
model_1.fit(X_train, y_train)
```

```python
Random_Prediction = model_1.predict(X_test)
Random_Prediction
```

```python
score_2= accuracy_score(y_test, Random_Prediction)
score_2
```

```python
#Comparing the accuracy of the two machine learning models
print(f"Decision Tree Accuracy: {score:.2f}")
print(f"Random Forest Accuracy: {score_2:.2f}")
```

In conclusion, since 0.78 > 0.75, Random Forest model performed better than the Decision Tree model. Therefore, we can conclude that the random forest model is the best model for this dataset.
